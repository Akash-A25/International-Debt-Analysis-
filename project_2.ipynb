```python
import matplotlib.pyplot as plt
df_long['Series Code'].value_counts().head(10).plot(kind = 'bar')
plt.title('top 10 count value')
plt.xlabel('Series_Code value')
plt.ylabel('Count')
plt.show()
```

```python
top10 =[
    df_long.groupby('Country Name')['Value']
    .sum()
    .nlargest(10)   
]
```

```python
plt.figure(figsize=(10,6))
plt.bar(top10[0].index, top10[0].values)
plt.xlabel('Country Name')
plt.ylabel('Total Value')
plt.title('Top 10 Countries by Total Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
lowest10 =[
    df_long.groupby('Country Name')['Value']
    .sum()
    .nsmallest(10)  
]
```

```python
plt.figure(figsize=(10,6))
lowest10[0].plot(kind='bar', color='lightcoral')
plt.title('Lowest 10 Countries by Total Value')
plt.xlabel('Country Name')  
plt.ylabel(' Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
yearly_avg = df_long.groupby('Year')['Value'].sum()

plt.figure(figsize=(10,6))
yearly_avg.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearly Average Value')
plt.show()
```

```python
Series_top10 =(
    df_long.groupby('Series Name')['Value']
    .sum()
    .nlargest(10)
)
print(Series_top10)
```

```python
plt.figure(figsize=(10,6))
Series_top10.plot(kind='bar', color='purple')
plt.title('Series_top10')
plt.xlabel('Series Name')
plt.ylabel('Value')   
plt.xticks(rotation=45, ha='right')
plt.show()
```
