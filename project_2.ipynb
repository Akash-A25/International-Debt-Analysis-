```python
#-----Read Data set 1 in pandas-------
data = pd.read_csv('D:\project_2\IDS_ALLCountries_Data.csv', encoding='latin-1')
```

```python
df_long
```

```python
df_long.isnull().sum()
```

```python
#------------ fill missing Series Name based on Country Code  --------
df_long['Series Name'] = df_long.groupby('Series Code')['Series Name'].bfill().ffill()
```

```python
#--------- fill missing Series Code based on Series Name   --------
df_long['Series Code'] = df_long.groupby('Series Name')['Series Code'].bfill().ffill()
```

```python
import matplotlib.pyplot as plt
#value based on series
df_long['Series Code'].value_counts().head(10).plot(kind = 'bar')
plt.title('top 10 count value')
plt.xlabel('Series_Code value')
plt.ylabel('Count')
plt.show()
```

```python
# values based on top country
top10 =[
    df_long.groupby('Country Name')['Value']
    .sum()
    .nlargest(10)   
]
print(top10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top10[0].index, top10[0].values)
plt.xlabel('Country Name')
plt.ylabel('Total Value')
plt.title('Top 10 Countries by Total Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#value based on lowest country
lowest10 =[
    df_long.groupby('Country Name')['Value']
    .sum()
    .nsmallest(10)  
]
print(lowest10)
```

```python
plt.figure(figsize=(10,6))
lowest10[0].plot(kind='bar', color='lightcoral')
plt.title('Lowest 10 Countries by Total Value')
plt.xlabel('Country Name')  
plt.ylabel(' Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# year wise average
yearly_avg = df_long.groupby('Year')['Value'].sum()

plt.figure(figsize=(10,6))
yearly_avg.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearly Average Value')
plt.show()
```

```python
#-----db connection----------
from sqlalchemy import create_engine,text

# Replace with your connection details
host = "localhost"
port = "5432"  
database = "inter"
username = "postgres"
password = "123AKA123"

# Create the connection string (URL format) -> postgres
engine_string = f"postgresql://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)


```

```python
df_long.head(1)
```

```python
#---create tables1 in sql -----
with engine.connect() as connection:
    connection.execute(
    text(""" create table ALLCountries_Data(
         Country_Name VARCHAR(100),
         Country_Code VARCHAR(50),
         Counterpart_Area_Name VARCHAR(50),
         Counterpart_Area_Code VARCHAR(15),
         Series_Name VARCHAR(300),
         Series_Code VARCHAR(50),
         Year INT,
         Value FLOAT
         )
    """))

connection.commit()
print("table create successfull")
```
